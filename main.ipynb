```python
from matplotlib import pyplot as plt 
import pandas as pd
from sklearn.cluster import KMeans 
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('income.csv')
data
```

```python
plt.scatter(data['Age']  , data['Income($)'])
plt.show()
```

```python
km = KMeans(n_clusters=3) # 3 cluster is required 
```

```python
y_predict = km.fit_predict(data[['Age' , 'Income($)']]) #fit and predict both
y_predict
```

```python
data['Cluster'] = y_predict
data.head()
```

```python
#defines 3- cluser in diffrentt
df1 = data[data.Cluster == 0]
df2 = data[data.Cluster == 1]
df3 = data[data.Cluster == 2]
print(df1)
print(df2)
label = "Income($)" 
plt.scatter(df1.Age , df1['Income($)'] , color = 'Blue' , label = label)
plt.scatter(df2.Age , df2['Income($)'] , color = 'Black', label = label)
plt.scatter(df3.Age , df3['Income($)'] , color = 'Red' , label = label)

plt.xlabel("Age")
plt.ylabel("Income")
plt.legend()
plt.show()
```

HERE SCALE IS NOT PERFECT Y-axis IS SOOO BIG ANG X-axis IS SOO SMALL SO WE HAVE TO
SOLVE THAT PROBLEM AND ARRANGE THE "RED" & "BLACK DOT IN PERFECT GROUP (Conver into scaler)

```python
sclaer = MinMaxScaler()

sclaer.fit(data[['Income($)']])
data['Income($)'] = sclaer.transform(data[['Income($)']])
 #we have income in 0 & 1 fomate

#now Age 
sclaer.fit(data[['Age']])
data['Age'] = sclaer.transform(data[['Age']])
data
```

```python
km = KMeans(n_clusters= 3)
y_predict_1 = km.fit_predict(data[['Age' , 'Income($)']])
y_predict_1
```

```python
data['Cluster'] = y_predict_1
data
```

```python
df1_new = data[data.Cluster == 0]
df2_new = data[data.Cluster == 1]
df3_new = data[data.Cluster == 2]

label = ['Income']
plt.scatter(df1_new.Age , df1_new['Income($)'] , color = '#4F709C' , label = label)
plt.scatter(df2_new.Age , df2_new['Income($)'],  color = '#040D12' , label = label)
plt.scatter(df3_new.Age , df3_new['Income($)'] , color = '#451952' , label = label)
plt.legend()
plt.show()
```

Make center using km.cluster_center

```python
km.cluster_centers_
```

```python
df1_new = data[data.Cluster == 0]
df2_new = data[data.Cluster == 1]
df3_new = data[data.Cluster == 2]

label = ['Income']
plt.scatter(df1_new.Age , df1_new['Income($)'] , color = '#4F709C' , label = label)
plt.scatter(df2_new.Age , df2_new['Income($)'],  color = '#040D12' , label = label)
plt.scatter(df3_new.Age , df3_new['Income($)'] , color = '#451952' , label = label)
plt.scatter(km.cluster_centers_[:,0] , km.cluster_centers_[:,1] , color = '#252B48' , marker="*" , label = "Center")
plt.legend()
plt.show()
```

now ALBO method
#fing the SSE

```python
k_range = range(1,11)
SSE = []
for k in k_range:
    km = KMeans(n_clusters= k)
    km.fit(data[['Age' , 'Income($)']])
    SSE.append(km.inertia_) #inertia_ --> for finding SSE 
```

```python
SSE #all sum of sqaure error
```

```python
plt.title("ALBO TECH..")
plt.xlabel("Range")
plt.ylabel("SSE")
plt.plot(k_range , SSE , marker = 'o')
plt.show()
```
